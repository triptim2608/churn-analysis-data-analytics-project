joining tables using aggregation to reduce duplication on particular account id

```python
usage_agg = usage.groupby('subscription_id').agg({
    'usage_count': 'sum',
    'usage_duration_secs': 'sum',
    'error_count': 'sum'
}).reset_index()
```

```python
subs_usage = subs.merge(usage_agg, on='subscription_id', how='left')
```

```python
df_main = accounts.merge(
    subs_usage,
    on='account_id',
    how='left'
)
```

```python
tickets.columns
```

```python
df_support_agg = tickets.groupby('account_id', as_index=False).agg({
    'ticket_id': 'count',
    'resolution_time_hours': 'mean',
    'satisfaction_score': 'mean'})


df_support_agg.rename(columns={
    'ticket_id': 'total_tickets',
    'resolution_time_hours': 'avg_resolution_time',
    'satisfaction_score': 'avg_satisfaction'}, inplace=True)
```

```python
df_main = pd.merge(df_main,df_support_agg,on='account_id', how='left')
```

```python
df_churn_clean = churn.sort_values('churn_date').drop_duplicates(subset='account_id',keep='last')
```

```python
df_churn_clean = df_churn_clean[['account_id','reason_code','feedback_text']]
```

```python
df_churn_clean.rename(columns={'reason_code': 'reason','feedback_text': 'feedback'}, inplace=True)
```

```python
df_main = pd.merge(df_main,df_churn_clean,on='account_id',how='left')
```

```python
churn.columns
```

```python
df_main['usage_count'] = df_main['usage_count'].fillna(0)
df_main['error_count'] = df_main['error_count'].fillna(0)
df_main['total_tickets'] = df_main['total_tickets'].fillna(0)

df_main['reason'] = df_main['reason'].fillna('Unknown')
df_main['feedback'] = df_main['feedback'].fillna('No Feedback')
```

```python
df_main.columns
```

```python
df_main['account_id'].value_counts().head()
```

```python
df_final = df_main.groupby('account_id', as_index=False).agg({
    'account_name': 'first',
    'industry': 'first',
    'country': 'first',
    'plan_tier_x': 'first',
    
    'usage_count': 'sum',
    'error_count': 'sum',
    'total_tickets': 'sum',
    
    'avg_resolution_time': 'mean',
    'avg_satisfaction': 'mean',
    
    'mrr_amount': 'mean',
    
    'reason': 'first',
    'feedback': 'first',
    
    'churn_flag_x': 'max'})
```

```python
df_final.rename(columns={'plan_tier_x': 'plan_tier','churn_flag_x': 'churn_flag'}, inplace=True)
```
